# Plot the histograms to have a quick look at the variable Distribution for all Numerical Variable

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
def diagnostic_plots(data, variable):
    
    # function to plot a histogram and a Q-Q plot
    # side by side, for a certain variable
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    data[variable].hist()

    plt.subplot(1, 2, 2)
    stats.probplot(data[variable], dist="norm", plot=plt)

    plt.show()
```

```python
#here i m checking for the normality of numerical values before encoding i applied , non of the numerical variable is normaly distributed 
for i in numerical:
    diagnostic_plots(data, i)
# Most of the variables are normally distributed  
```
